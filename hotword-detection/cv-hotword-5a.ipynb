## 1. Data Procecssing

```python
FT_ASR_MODEL = r'C:\Users\jared\OneDrive\Documents\VSCode\job\htx\ds-test\asr-train\wav2vec2-large-960h-cv'
CSV_PATH_TASK_2A = r"C:\Users\jared\OneDrive\Documents\VSCode\job\htx\ds-test\asr\cv-valid-dev.csv"
```

```python
from transformers import AutoModelForCTC

# loading for inference
ft_model = AutoModelForCTC.from_pretrained(FT_ASR_MODEL)
```

```python
# Put the model in evaluation mode
ft_model.eval() 
```

```python
import pandas as pd

all_train_results = pd.read_csv(CSV_PATH_TASK_2A)
all_train_results
```

```python
selected_columns = ['filename', 'text']

train_df = all_train_results[selected_columns]
train_df
```

```python
from transformers import  Wav2Vec2Processor

# Create processor pipeline (Feature extractor built into processor)
processor = Wav2Vec2Processor.from_pretrained("facebook/wav2vec2-large-960h")
```

```python
import librosa
import re
import os

chars_to_ignore_regex = '[\,\?\.\!\-\;\:\"]'

def preprocess_audio(audio_dir, filename):    
    print(audio_dir)
    print(filename)
    full_path = os.path.join(audio_dir, filename) 
    print(f"Loading audio from: {full_path}")  # Print the full path
    try:
        audio, sample_rate = librosa.load(full_path, sr=16000) 
        print(audio)
        print(sample_rate)
        
        # Checking if resampling if necessary
        if sample_rate != 16000:
            audio = librosa.resample(audio, orig_sr=sample_rate, target_sr=16000) 
        
    except Exception as e:
        print("Failed to transcribe audio" + str(e))

    return audio

def remove_special_characters(batch):
    batch["text"] = re.sub(chars_to_ignore_regex, '', batch["text"]).upper()
    return batch

def prepare_dataset(batch):
    audio = batch["audio"]
    batch["input_values"] = processor(audio, sampling_rate=16000).input_values[0]
    
    with processor.as_target_processor():
        batch["labels"] = processor(batch["text"]).input_ids
    return batch
```

```python
AUDIO_DIR = r"C:\Users\jared\Downloads\common_voice\cv-valid-dev"
```

```python
train_df['audio'] = train_df['filename'].map(lambda x: preprocess_audio(AUDIO_DIR, x))
train_df
```

```python
from datasets import Dataset


# Convert the pandas DataFrame to a Hugging Face Dataset
train_ds = Dataset.from_pandas(train_df) 
train_ds = train_ds.map(remove_special_characters)
train_ds = train_ds.map(prepare_dataset) 
train_ds
```

```python
import torch

def inference_result(batch):
  with torch.no_grad():
    input_values = torch.tensor(batch["input_values"]).unsqueeze(0)
    logits = ft_model(input_values).logits

  pred_ids = torch.argmax(logits, dim=-1)
  batch["pred_str"] = processor.batch_decode(pred_ids)[0]
  batch["text"] = processor.decode(batch["labels"], group_tokens=False)
  
  return batch
```

```python
train_results = train_ds.map(inference_result)
train_results
```

```python
idx = 47
print(f"Text: {train_results[idx]['text']}")
print(f"Prediction: {train_results[idx]['pred_str']}") 
```

```python
selected_columns = ['filename', 'text', 'pred_str']

all_results_df = train_results.to_pandas() 
results_df = all_results_df[selected_columns]
results_df
```

```python
results_df.to_csv("train_results_5a.csv")
```

## 2. Hotword Detection

```python
# converting to lower case
results_df['pred_str_lowercase'] = results_df['pred_str'].str.lower()
results_df
```

```python
# Define hotwords
HOTWORDS = ["be careful", "destroy", "stranger"]

def contains_hotword(text):
  for hotword in HOTWORDS:
    if hotword.lower() in text.lower():
      return 1
  return 0
```

```python
# Create a new column 'contains_hotword'

results_df['contains_hotword'] = results_df['pred_str_lowercase'].apply(lambda x: contains_hotword(x))
results_df
```

```python
# Get filenames of rows with hotwords
detected_files = results_df[results_df['contains_hotword'] == 1]['filename'] 
detected_files
```

```python
# Save filenames to detected.txt
with open("detected.txt", "w") as f:
    for filename in detected_files:
        f.write(f"{filename}\n")
```
